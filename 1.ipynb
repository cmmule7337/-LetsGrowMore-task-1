```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Iris.csv")
df
```

```python
df1 = df.loc[df['species'] == "Iris-virginica"]
df1
```

```python
temp = pd.DataFrame(df1.petal_length.value_counts())
temp.reset_index(level=0, inplace=True)
temp = temp.rename(columns={"index": "Petal Length", "petal_length": "Frequency"})
temp
```

```python
temp['Probability'] = temp['Frequency'] / sum(temp['Frequency'])
temp
```

```python
expected = sum(temp['Petal Length'] * temp['Probability'])
print("Expected Value :", expected)
```
