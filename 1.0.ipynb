```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Iris.csv")
df
```

```python
df1 = df.loc[df['species'] == "Iris-setosa"]
df1
```

```python
total_count = np.count_nonzero(df1.sepal_length)
total_count
```

```python
temp = df1.sepal_length.unique()
temp
```

```python
frequency = []
probability = []
count = 0

for i in temp:
    for j in df1["sepal_length"]:
        if i == j:
            count += 1
    frequency.append(count)
    probability.append(count/total_count)
    count = 0
    
print(temp)
print(frequency)
print(probability)
```

```python
df2 = pd.DataFrame({'Sepal Length' : temp, 'Frequency' : frequency, 'Probability' : probability})
df2
```

```python
expect_value = 0
i = 0
for value in temp:
    expect_value += (value * probability[i])
    i += 1
print('Expected Value =', expect_value)
```
